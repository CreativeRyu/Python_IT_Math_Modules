## **Kommutativgesetz**

### ***a + b = b + a***
### ***a * b = b * a***

```python
# Kommutativgesetz
a = 0
b = 1

# Addition => Oder
print(a or b)
print(b or a)
print((a or b) == (b or a))

# Multiplikation => Und
print(a and b)
print(b and a)
print((a and b) == (b and a))
```

## **Assoziativgesetz**

### ***(a + b) + c = a + (b + c)***
### ***(a * b) * c = a * (b * c)***

```python
# Assoziativgesetz
a = 1
b = 1
c = 1

# Addition => Oder
print(a or (b or c))
print((a or b) or c)
print((a or (b or c)) == ((a or b) or c))

# Multiplikation => Und
print(a and (b and c))
print((a and b) and c)
print((a and (b and c)) == ((a and b) and c))
```

## **Distributivgesetz**

### ***(a + b) * c = a * c + b * c***
### ***a * (b + c) = a * b + a * c***

```python
# Distributivgesetz
a = 1
b = 0
c = 1

print((a or b) and c)
print(a and c or b and c)
print(((a or b) and c) == (a and c or b and c))

print(a and (b or c))
print(a and b or a and c)
print((a and (b or c)) == (a and b or a and c))
```

## **Neutralittsgesetz**

### ***A  1 == A***
### ***A  0 == A***


```python
a = 1
b = 0

print(a and 1 == a)
print(b and 1 == b)

print(a or 0 == a)
print(b or 0 == b)
```
