## **Konjunktion und Disjunktion**
in Python mit "and" und "or"
+   "and" entspricht der Multiplikation
+   "or" entspricht der Addition



#### **Konjunktion in Python**
*If we have some free time, and the weather is good, we will go for a walk.*

```python
A = True
B = True

print("A: True & B: True")
print(A and B)

#############################

A = True
B = False

print("A: True & B: False")
print(A and B)

#############################

A = False
B = True

print("A: False & B: True")
print(A and B)

#############################

A = False
B = False

print("A: False & B: False")
print(A and B)
```

#### **Disjunktion in Python**
*If you are in the mall or I am in the mall, one of us will buy a gift for Mom.*

```python
A = True
B = True

print("A: True or B: True")
print(A or B)

#############################

A = True
B = False

print("A: True or B: False")
print(A or B)

#############################

A = False
B = True

print("A: False or B: True")
print(A or B)

#############################

A = False
B = False

print("A: False or B: False")
print(A or B)
```

## **Implikation**
*Wenn es regnet, ist die Strae nass.*\
*A impliziert B*

    A = es regnet
    B = die Strae ist nass



```python
A = True
B = True

print(not(A) or (A and B)) # A  (A  B)
print(not(A) or B) # A  B
```

## **quivalenz**
*Die Aussage ist nur dann Wahr wenn A und B gleich sind*

    A  B
    (A  B)  (B  A)
    (A  B)  (B  A)

+ A impliziert B
+ B impliziert A



```python
A = False
B = False

print((not(A) or B) and (not(B) or A))
```

## **Exklusiv ODER XOR**

    A  B
    (A  B)



```python
A = True
B = True

print(not((not(A) or B) and (not(B) or A)))
```
