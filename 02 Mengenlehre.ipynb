*Eine Menge ist die Zusammenfassung von beliebigen Objekten aller Art in einem Ganzen*\
*Diese Objekten können wirklich existieren oder gedacht sein.*

x ∈ M => x ist ein Element von M -> ist also Teil der Menge M\
x ∉ M => x ist kein Element von M -> ist also nicht in der Menge M enthalten\
{} oder ∅ ist eine leere Menge


Natürliche Zahlen ℕ\
Ganze Zahlen ℤ\
Rationale Zahlen ℚ\
Reelle Zahlen ℝ

Beispiel:\
43 ∈ ℕ
3.147 ∉ ℕ



```python
# Eine Menge ist in Python ein Set
# Elemente sind entweder in einer Menge oder nicht
# Es gibt keine Anzahl an Elementen in einer Menge

haustiere = {"Fynn", "Felix", 4, 7, 4, 7}

print(haustiere)

# auch mit der Length Funktion werden keine Redundanzen berücksichtigt
print(len(haustiere))
```

## **Teilmenge**
*Eigentliche oder Echte Teilmenge*

Definition:

A heißt Teilmenge von B\
A ⊂ B\
genau dann, wenn aus x ∈ A auch x ∈ B folgt.\
Ausgeschrieben: x ∈ A => x ∈ B.



B ⊊ A
gilt dann, wenn ein x ∈ B existiert, sodass x ∉ A.

=> sprich B ist keine Teilmenge von A, sobald ein Element in B enthalten ist,\
dass sich nicht auch in der Menge A befindet.
⊂⊊⊆⊇⊋⊃
