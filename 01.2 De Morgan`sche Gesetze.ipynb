# **1.2 De Morgan`sche Gesetze**

## ***¬(A ^ B) = ¬A v ¬ B***
## ***¬(A v B) = ¬A ^ ¬B***

```python
A = True
B = False

print(not(A and B))
print(not(A) or not(B))

print(not(A or B))
print(not(A) and not(B))
```

```python
A = True
B = True

print(not((not(A) or B) and (not(B) or A)))

# De Morgan'sche Gesetze
print((not(not(A) or B) or not((not(B) or A))))

print(((A and not(B)) or (B and not(A))))
```
